```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("prevalence-by-mental-and-substance-use-disorder.csv")
df2 = pd.read_csv("mental-and-substance-use-as-share-of-disease.csv")
```

<h3>Exploratory Data Analysis</h3>

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2)
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop("Code", axis=1, inplace=True)
```

```python
df.head()
```

```python
cols = ["Country","Year","Schizophrenia","Bipolar_disorder","Eating_disorders","Anxiety_disorders","Drug_use_disorders",
        "Depressive_disorders","Alcohol_use_disorders","Mental_Fitness"]

df.columns = cols
```

```python
df.head()
```

```python
sns.pairplot(df, corner=True)
plt.plot()
```

```python
mean = df["Mental_Fitness"].mean()
mean
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in df.columns:
    if df[i].dtype == "object":
        df[i] = le.fit_transform(df[i])
```

```python
df.shape
```

```python
df.head
```

<h3>Splitting Data into Training dataset and Testing dataset</h3>

```python
X = df.drop("Mental_Fitness", axis=1)
y = df["Mental_Fitness"]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=.20, random_state=2)
```

```python
print(xtrain.shape, xtest.shape)
print(ytrain.shape, ytest.shape)
```

<h3>Linear Regression Model</h3>

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
linreg = LinearRegression()
linreg.fit(xtrain, ytrain)
```

```python
ytrain_pred = linreg.predict(xtrain)
```

```python
mse = mean_squared_error(ytrain, ytrain_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2 = r2_score(ytrain, ytrain_pred)
r2
```

```python
print("The Resultant performance of the linear model is given below: ")
print("Mean Squared error: {}". format(mse))
print("Root Mean Squared error: {}". format(rmse))
print("R-squared error: {}". format(r2))
```

<h3>Random Forest Regressor</h3>

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
ranfor = RandomForestRegressor()
ranfor.fit(xtrain, ytrain)
```

```python
ytrain_pred_rf = ranfor.predict(xtrain)
```

```python
mse_rf = mean_squared_error(ytrain, ytrain_pred_rf)
mse_rf
```

```python
rmse_rf = np.sqrt(mse_rf)
rmse_rf
```

```python
r2_rf = r2_score(ytrain, ytrain_pred_rf)
r2_rf
```

```python
print("The Resultant performance of the Random Forest Regressor is given below: ")
print("Mean Squared error: {}". format(mse_rf))
print("Root Mean Squared error: {}". format(rmse_rf))
print("R-squared error: {}". format(r2_rf))
```

<h3>Linear Regression Evaluation</h3>

```python
ytest_pred = linreg.predict(xtest)
```

```python
mse_pred_lin = mean_squared_error(ytest, ytest_pred)
mse_pred_lin
```

```python
rmse_pred_lin = np.sqrt(mse_pred_lin)
rmse_pred_lin
```

```python
r2_pred_lin = r2_score(ytest, ytest_pred)
r2_pred_lin
```

```python
print("The performance of Linear Regression model after evaluation is: ")
print("Mean Squared error: {}". format(mse_pred_lin))
print("Root Mean Squared error: {}". format(rmse_pred_lin))
print("R-squared error: {}". format(r2_pred_lin))
```

<h3>Random Forest Regressor Evaluation</h3>

```python
ytest_pred_rf = ranfor.predict(xtest)
```

```python
mse_pred_rf = mean_squared_error(ytest, ytest_pred_rf)
mse_pred_rf
```

```python
rmse_pred_rf = np.sqrt(mse_pred_rf)
rmse_pred_rf
```

```python
r2_pred_rf = r2_score(ytest, ytest_pred_rf)
r2_pred_rf
```

```python
print("The performance of Random Forest Regressor model after evaluation is: ")
print("Mean Squared error: {}". format(mse_pred_rf))
print("Root Mean Squared error: {}". format(rmse_pred_rf))
print("R-squared error: {}". format(r2_pred_rf))
```

<h2>Take away point</h2>

Random Forest Model outperforms Linear Regression Model in both the Training and Testing data
